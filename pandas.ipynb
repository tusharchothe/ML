## 1. Series

```python
x =[1,2,3,4]
var = pd.Series(x,index=['a','b','c','d'],dtype='float',name='python')

print(var)
print(type(var))
print(var[2])

print()

# for Dictionary
dic={'name':['python','c','c++'],'popularity':[16,32,22],'rank':[1,3,2]}
var1=pd.Series(dic)
print(var1)
```

```python
s = pd.Series(12)
print(type(s),s)

# with index
s = pd.Series(12,index=[1,2,3,4,5,6])
print(type(s),s)

s1 = pd.Series(12,index=[1,2,3,4,5,6,7,8,9])
s2 = pd.Series(12,index=[1,2,3,4])

print(s1+s2)
# benefit of pandas  over numpy is we can perform operation on missing data
# and in numpy if try it shows the broadcasting error 
```

## 9.Merge Function

```python
var1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
var2=pd.DataFrame({'A':[1,2,3],"C":[44,55,66]})

pd.merge(var1,var2,on="A")
```

```python
# how parameter= handling ,missing data in different dataframes

#left
var1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
var2=pd.DataFrame({'A':[1,2,5],"C":[44,55,66]})

pd.merge(var1,var2,how='left')
```

```python
# right
var1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
var2=pd.DataFrame({'A':[1,2,5],"C":[44,55,66]})

pd.merge(var1,var2,how='right')
```

```python
# outer
var1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
var2=pd.DataFrame({'A':[1,2,5],"C":[44,55,66]})

pd.merge(var1,var2,how='outer',indicator=True)
```

```python
var1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
var2=pd.DataFrame({'A':[1,2,5],'B':[44,55,66]})

pd.merge(var1,var2,left_index=True,right_index=True,suffixes=('name','id'))
```

## 12.join and append function 

```python
var1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]},index=['a','b','c'])
var2=pd.DataFrame({'C':[7,8,9],'D':[10,11,12]},index=['a','b','c'])

var2.join(var2,how='inner',lsuffix='_11',rsuffix='_12')
```

```python
var1.join(var2)
```

```python
# Append
var1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
var2=pd.DataFrame({'C':[11,22,33,44,55],'D':[44,55,66,77,88]})

var1.append(var2)
# it is removed from pandas instead of this use concat function 
```

```python
import pandas as pd

var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'C': [11, 22, 33, 44, 55], 'D': [44, 55, 66, 77, 88]})

result = pd.concat([var1, var2], axis=0)  # axis=0 means stacking vertically
print(result)
```
